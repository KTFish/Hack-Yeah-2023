```python
df_pivot = df_work.iloc[:4, :18]
df_pivot = df_pivot.T.reset_index(drop=False).T
df_pivot.columns = df_pivot.iloc[1]
df_pivot = df_pivot.iloc[:1].append(df_pivot.iloc[2:])
skip_rows = 3
# get information about year and languages
tmp_row = list(df_pivot.iloc[0])
languages = []
languages = tmp_row[skip_rows::5]
print(languages)

# remove language row
df_pivot = df_pivot.iloc[1:]

# duplicate X times each row
x = len(languages)
df_pivot = df_pivot.loc[np.repeat(df_pivot.index.values, x)]


df_pivot.reset_index(drop=True, inplace=True)

# fill values for each language row
x_languages = len(languages)
index = 0
for i in range(len(df_pivot)):
    if i % x_languages == 0:
        index = i
    else:
        df_pivot.iloc[i][skip_rows:(skip_rows+5)] = df_pivot.iloc[index][skip_rows+5*(i%x_languages):(skip_rows+5)+5*(i%x_languages)]
    
# remove unnecessary columns
df_pivot = df_pivot.iloc[:, :8]

# fill informations about languages
df_pivot["language"] = languages * (len(df_pivot) // len(languages))

# extract school number
df_pivot['Nr szkoły'] = df_pivot['Nazwa szkoły'].str.extract(r'NR (\d+)')
df_pivot

# df_pivot.to_csv('Staniny_2_2022_v2.csv', index=False)
```
